```python
import pandas as pd
import numpy as np
import nltk
import re
from nltk.corpus import stopwords
import matplotlib.pyplot as ply
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

```

```python
#data preparation
df = pd.read_csv('spam.csv', encoding='latin1')
df.head()
```

```python
df = df[['v2','v1']]
df = df.rename(columns={'v2': 'messages','v1':'label'})
df.head()
```

```python
df.isnull().sum()
```

```python
STOPWORDS = set(stopwords.words('english'))

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^0-9a-zA-Z]', '  ', text)
    text = re.sub(r'\s+','',text)
    text = " ".join(word for word in text.split() if word not in STOPWORDS)
    return text
```

```python
#Model training
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, TfidfTransformer

def classify(model, X, y):
    # train test split
    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, shuffle=True, stratify=y)
    # model training
    pipeline_model = Pipeline([('vect', CountVectorizer()),
                               ('tfidf',TfidfTransformer()),
                               ('clf', model)])
    pipeline_model.fit(x_train, y_train)
    
    print('Accuracy:', pipeline_model.score(x_test, y_test)*100)
    
#     cv_score = cross_val_score(model, X, y, cv=5)
#     print("CV Score:", np.mean(cv_score)*100)
    y_pred = pipeline_model.predict(x_test)
    print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
classify(model, X, y)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
classify(model, X, y)
```

```python
from sklearn.svm import SVC
model = SVC(C=3)
classify(model, X, y)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
classify(model, X, y)
```

```python
#Prediction Script
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.naive_bayes import MultinomialNB

pipeline_model = Pipeline([
    ('vect', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('clf', MultinomialNB())
])

X = df['cleaned_message']  
y = df['label']            
pipeline_model.fit(X, y)


def predict_category(text):
    """
    Input: text (string)
    Output: predicted label
    """
    cleaned_text = clean_text(text)
    prediction = pipeline_model.predict([cleaned_text])
    return prediction[0]


print("Enter a sentence to predict its category (type 'exit' to quit):\n")

while True:
    user_input = input("Your sentence: ")
    if user_input.lower() == 'exit':
        print("Exiting prediction script.")
        break
    elif user_input.strip() == '':
        print("Please enter a valid sentence.")
    else:
        label = predict_category(user_input)
        print(f"Predicted category: {label}\n")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, shuffle=True, stratify=y
)
```

```python
#Small visualization
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

# Dictionary to store models and predictions
models = {
    'Naive Bayes': MultinomialNB(),
    'Logistic Regression': LogisticRegression(max_iter=200),
    'SVM': SVC(C=3),
    'Random Forest': RandomForestClassifier()
}

predictions = {}

for name, model in models.items():
    pipeline = Pipeline([
        ('vect', CountVectorizer()),
        ('tfidf', TfidfTransformer()),
        ('clf', model)
    ])
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    predictions[name] = y_pred
```

```python
#Confusion matrix
def plot_confusion(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred, labels=y.unique())
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=y.unique(),
                yticklabels=y.unique())
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(title)
    plt.show()

for name, y_pred in predictions.items():
    plot_confusion(y_test, y_pred, f'Confusion Matrix - {name}')
```
